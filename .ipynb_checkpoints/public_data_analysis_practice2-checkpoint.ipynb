### 구별 학원업종 분석

```python
df_academy = df.loc[df['상권업종대분류명'] == '학문/교육']
df_academy
```

```python
# 시군구명과 상권업종중분류명으로 그룹화
df_academy_count = df_academy.groupby(['시군구명','상권업종중분류명'])['상호명'].count().copy()
```

```python
df_academy_count
```

```python
df_academy_count = df_academy_count.reset_index()
```

```python
df_academy_count = df_academy_count.rename(columns = {'상호명' : '상호수'})
```

```python
df_academy_count.head()
```

```python
# 시군구명과 상권업종소분류명으로 그룹화
df_academy_count2 = df_academy.groupby(['시군구명','상권업종소분류명'])['상호명'].count()
```

```python
df_academy_count2 = df_academy_count2.reset_index()
```

```python
df_academy_count2 = df_academy_count2.rename(columns = {'상호명' : '상호수'})
```

```python
df_academy_count2
```

```python
# df_academy_count2에서 빈도수가 1000이상인 데이터만 가져오기

df_academy_count2_top100 = df_academy_count2[df_academy_count2['상호수'] > 100]
```

## 데이터 전처리

### 필요없는 컬럼 제거하기

```python
# 결측치가 있는 컬럼은 제거하기

df_null = df.isnull().sum().sort_values(ascending = False).head(9)
```

```python
# 코드로 이루어진 컬럼은 제거
cols = df.columns
```

```python
not_use_cols = cols[cols.str.contains('코드|번호')]
```

```python
df = df.drop(not_use_cols, axis = 1)
```

```python
df.head()
```

### 상호명 전처리

```python
df['상호명_소문자'] = df['상호명'].str.lower()
```

## 데이터 분석(스타벅스, 맥도날드)

### 스타벅스, 맥도날드 데이터만 가져오기

```python
# 상호명 중 NaN값이 있는 행은 제외
df = df[~df['상호명'].isnull()]
```

```python
df_star_mc = df[df['상호명_소문자'].str.contains('스타벅스|starbucks|맥도날드|멕도날드|mcdonal')].copy()
```

```python
df_star_mc.info()
```

```python
df_star_mc['상호명_소문자'].unique()
```

```python
# 상호명 통일시키기(스타벅스, 맥도날드)
df_star_mc.loc[df_star_mc['상호명_소문자'].str.contains('스타벅스|starbucks'),'브랜드명'] = '스타벅스'
```

```python
df_star_mc.loc[df_star_mc['상호명_소문자'].str.contains('맥도날드|멕도날드|mcdonal'), '브랜드명'] = '맥도날드'
```

```python
df_star_mc['브랜드명'].shape
```

```python
df_star_mc['상권업종대분류명'].value_counts()
```

```python
df_star_mc[(df_star_mc['브랜드명'] == '스타벅스') & (df_star_mc['시군구명'] == '송파구')]
```

```python
# 시군구별로 맥도날드와 스타벅스가 몇개씩 있는지 확인
df_count = df_star_mc.groupby(['시군구명','브랜드명'])['상호명'].count().reset_index().copy()
```

```python
df_count.set_index('시군구명')
```

### 데이터 시각화

```python
plt.figure(figsize = (15,5))
graph = sns.countplot(data = df_star_mc, x = '시군구명', hue = '브랜드명')
graph.text()
```

```python

sns.catplot(data = df_count, x = '브랜드명', y = '상호명', kind = 'bar', col = '시군구명', hue = '시군구명', col_wrap = 4, sharex = False)
```

### 지도에 표시하기

```python
# 라이브러리 가져오기
import folium
```

```python
# 지도의 중심 위도,경도 설정
lat = df_star_mc['위도'].mean()
long = df_star_mc['경도'].mean()
```

```python
m = folium.Map(location = [lat, long])

for i in df_star_mc.index:
    df_lat = df_star_mc.loc[i,'위도']
    df_long = df_star_mc.loc[i,'경도']
    df_brandname = df_star_mc.loc[i,'브랜드명']
    df_brandlocation = df_star_mc.loc[i,'도로명주소']
    
    if df_brandname == '스타벅스':
        icon_color = 'green'
    else:
        icon_color = 'red'
    
    folium.Marker([df_lat, df_long],  tooltip=df_brandname + '/' + df_brandlocation, icon = folium.Icon(color = icon_color)).add_to(m)
    
m
```
