## 데이터 전처리
### 결측치 처리

```python
null = df.isnull().sum()
null
```

```python
# 결측치가 많은 행들만 가져오기
not_use_col = null.sort_values(ascending = False).head(9)
not_use_col
```

```python
not_use_col.index
```

```python
# 결측치가 많은 컬럼은 데이터에서 삭제
print(df.shape)
df = df.drop(not_use_col.index, axis = 1)
print(df.shape)
```

```python
df.info()
```

### 사용하지 않는 컬럼들 제거하기

```python
cols = df.columns
```

```python
# 컬럼명에서 코드 or 번호가 있는 컬럼들만 선택

not_use_col2 = cols[cols.str.contains('코드|번호')]
```

```python
df = df.drop(not_use_col2, axis = 1)
```

```python
df.info()
```

```python
# 상호명에 결측치가 있는 행 제거
df = df.dropna()
```

## 데이터 분석

### 구별 음식점 업종 분석

```python
# 상권업종대분류명이 음식이 데이터 가져오기
df_seoul_food = df.loc[df['상권업종대분류명'] == '음식'].copy()
df_seoul_food.head()
```

```python
# 시군구명, 상권업종중분류명으로 그룹화하여 상점수 확인하기
df_seoul_food_grp = df_seoul_food.groupby(['시군구명','상권업종중분류명'])['상호명'].count().copy()
df_seoul_food_grp
```

### 구별 학원업종 분석

```python
df_academy = df.loc[df['상권업종대분류명'] == '학문/교육']
df_academy
```

```python
# 시군구명과 상권업종중분류명으로 그룹화
df_academy_count = df_academy.groupby(['시군구명','상권업종중분류명'])['상호명'].count().copy()
```

```python
df_academy_count
```

```python
df_academy_count = df_academy_count.reset_index()
```

```python
df_academy_count = df_academy_count.rename(columns = {'상호명' : '상호수'})
```

```python
df_academy_count.head()
```

```python
# 시군구명과 상권업종소분류명으로 그룹화
df_academy_count2 = df_academy.groupby(['시군구명','상권업종소분류명'])['상호명'].count()
```

```python
df_academy_count2 = df_academy_count2.reset_index()
```

```python
df_academy_count2 = df_academy_count2.rename(columns = {'상호명' : '상호수'})
```

```python
df_academy_count2
```

```python
# df_academy_count2에서 빈도수가 1000이상인 데이터만 가져오기

df_academy_count2_top100 = df_academy_count2[df_academy_count2['상호수'] > 100]
```

```python
df_academy_count2_top100.head()
```

### 구별 학원업종 시각화

```python
plt.figure(figsize = (15,3))
sns.barplot(data = df_academy_count, x = '시군구명', y = '상호수')
```

```python
# 시군구별 서브플롯 만들기
plt.figure(figsize = (15,3))
sns.catplot(data = df_academy_count2_top100, x = '상권업종소분류명', y = '상호수', kind = 'bar', col = '시군구명', hue = '시군구명', col_wrap = 1, sharex = False)
```

## 구별 학원업종 지도에 표시하기

```python
import folium
```

```python
# 상권업종소분류명에 검도도장, 합기도장이 들어가있는 데이터 가져오기
KH = df_academy[df_academy['상권업종소분류명'].isin(['검도도장','합기도장'])]
```

```python
KH
```

```python
lat = df_academy['위도'].mean()
long = df_academy['경도'].mean()
m = folium.Map(location = (lat, long), zoom_start = 12)
m
```

```python
folium.Marker(
    [37.532601,127.074243], tooltip='비상수학'
).add_to(m)
```

```python
len(KH)
```

```python
lat = df_academy['위도'].mean()
long = df_academy['경도'].mean()
m = folium.Map(location = (lat, long), zoom_start = 12)

for i in KH.index:
    lat_KH = KH.loc[i,'위도']
    long_KH = KH.loc[i,'경도']
    tooltip_KH = KH.loc[i,'상호명']
    folium.Marker([lat_KH,long_KH], tooltip=tooltip_KH).add_to(m)
    
m
```
